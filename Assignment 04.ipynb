```python
#1
class Customer:
  def __init__(self,name):
    self.name=name
  def greet(self,name=None):
    if name:
      print(f'Hello {self.name}!')
    else:
      print('Hello!')
  def purchase(self,*items):
    print(self.name+", you purchased "+str(len(items))+' items(s):')
    for i in items:
      print(i)
customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#9
class Batsman:
  def __init__(self,name=None,runsScored=None,ballsFaced=None):
    
    
    if ballsFaced:
      self.name=name
      self.runsScored=runsScored
      self.ballsFaced=ballsFaced   
    else:
      self.runsScored=name
      self.ballsFaced=runsScored
      self.name='New Batsman'

  def battingStrikeRate(self):    
    self.battingStrikeRate=self.runsScored/self.ballsFaced*100
    return self.battingStrikeRate
  def setName(self,name):
    self.name=name
  def printCareerStatistics(self):
    print('Name:', self.name)
    print(f'Runs Scored: {self.runsScored}, Balls Faced: {self.ballsFaced}')
b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#14
class StudentDatabase:
  def __init__(self,name,id):
    self.name=name
    self.id=id
    self.grades={}

  def calculateGPA(self,grade_list,semester):
    l=[]
    n=[]
    for i in grade_list:
      x=i.split(':')
      l.append(x[0])
      n.append(float(x[1]))
      sum=0
      count=0
      for j in n:
        sum+=j
        count+=1
       
    self.cgpa=sum/count 
    self.grades.update({semester:{tuple(l):self.cgpa}})

  def printDetails(self):
    print('Name:', self.name)
    print('ID:', self.id)
    for i in self.grades:

      print('Courses taken in', i)
      for k in self.grades[i]:
        for j in k:
          print(j)
      print('GPA:',self.cgpa)


s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 
'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 
'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
